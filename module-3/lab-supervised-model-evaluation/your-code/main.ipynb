## Regression Model Evaluation

```python
from sklearn.datasets import load_boston

X, y = load_boston(return_X_y=True)
```

## 1. Split this data set into training (80%) and testing (20%) sets.

The `MEDV` field represents the median value of owner-occupied homes (in $1000's) and is the target variable that we will want to predict.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## 2. Train a `LinearRegression` model on this data set and generate predictions on both the training and the testing set.

```python
from sklearn.linear_model import LinearRegression
boston_model = LinearRegression()
boston_model.fit(X_train, y_train)
y_train_pred = boston_model.predict(X_train)
y_test_pred = boston_model.predict(X_test)
```

## 3. Calculate and print R-squared for both the training and the testing set.

```python
from sklearn.metrics import r2_score
print(f"R^2 Score for the training set: {r2_score(y_train_pred, y_train)}")
print(f"R^2 Score for the test set: {r2_score(y_test_pred, y_test)}")
```

## 4. Calculate and print mean squared error for both the training and the testing set.

```python
from sklearn.metrics import mean_squared_error
print(f"MSE for the training set: {mean_squared_error(y_train_pred, y_train)}")
print(f"MSE for the test set: {mean_squared_error(y_test_pred, y_test)}")
```

## 5. Calculate and print mean absolute error for both the training and the testing set.

```python
from sklearn.metrics import mean_absolute_error
print(f"MAE for the training set: {mean_absolute_error(y_train_pred, y_train)}")
print(f"MAE for the test set: {mean_absolute_error(y_test_pred, y_test)}")
```

## Classification Model Evaluation

```python
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)
```

```python
print(load_iris()["DESCR"])
```

## 6. Split this data set into training (80%) and testing (20%) sets.

The `class` field represents the type of flower and is the target variable that we will want to predict.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## 7. Train a `LogisticRegression` model on this data set and generate predictions on both the training and the testing set.

```python
from sklearn.linear_model import LogisticRegression
iridis_model = LogisticRegression()
iridis_model.fit(X_train, y_train)
y_train_pred = iridis_model.predict(X_train)
y_test_pred = iridis_model.predict(X_test)
```

## 8. Calculate and print the accuracy score for both the training and the testing set.

```python
from sklearn.metrics import accuracy_score
print(f"Accuracy for the training set: {accuracy_score(y_train_pred, y_train)}")
print(f"Accuracy for the test set: {accuracy_score(y_test_pred, y_test)}")
```

## 9. Calculate and print the balanced accuracy score for both the training and the testing set.

```python
from sklearn.metrics import balanced_accuracy_score
print(f"Balanced accuracy for the training set: {balanced_accuracy_score(y_train_pred, y_train)}")
print(f"Balanced accuracy for the test set: {balanced_accuracy_score(y_test_pred, y_test)}")
```

## 10. Calculate and print the precision score for both the training and the testing set.

```python
from sklearn.metrics import precision_score
print(f"Precision score for each class of the training set: {precision_score(y_train_pred, y_train, average=None)}")
print(f"Precision score for each class of the test set: {precision_score(y_test_pred, y_test, average=None)}")
print(f"Precision score (weighted) for the training set: {precision_score(y_train_pred, y_train, average='weighted')}")
print(f"Precision score (weighted) for the test set: {precision_score(y_test_pred, y_test, average='weighted')}")
```

## 11. Calculate and print the recall score for both the training and the testing set.

```python
from sklearn.metrics import recall_score
print(f"Reacall score for each class of the training set: {recall_score(y_train_pred, y_train, average=None)}")
print(f"Reacall score for each class of the test set: {recall_score(y_test_pred, y_test, average=None)}")
print(f"Reacall score (weighted) for the training set: {recall_score(y_train_pred, y_train, average='weighted')}")
print(f"Reacall score (weighted) for the test set: {recall_score(y_test_pred, y_test, average='weighted')}")
```

## 12. Calculate and print the F1 score for both the training and the testing set.

```python
from sklearn.metrics import f1_score
print(f"F1 score for each class of the training set: {f1_score(y_train_pred, y_train, average=None)}")
print(f"F1 score for each class of the test set: {f1_score(y_test_pred, y_test, average=None)}")
print(f"F1 score (weighted) for the training set: {f1_score(y_train_pred, y_train, average='weighted')}")
print(f"F1 score (weighted) for the test set: {f1_score(y_test_pred, y_test, average='weighted')}")
```

## 13. Generate confusion matrices for both the training and the testing set.

```python
from sklearn.metrics import confusion_matrix
iris_labels = ["Setosa", "Versicolour", "Virginica"]
cm_train = confusion_matrix(y_train, y_train_pred)
cm_test = confusion_matrix(y_test, y_test_pred)
```

```python
from seaborn import heatmap
ax = heatmap(cm_train, annot=True, linewidths=.5)
ax.set(title="Confusion Matrix for the train set", xticklabels=iris_labels, yticklabels=iris_labels, xlabel="True Species", ylabel="Predicted Species")
```

```python
ax = heatmap(cm_test, annot=True, linewidths=.5)
ax.set(title="Confusion Matrix for the test set", xticklabels=iris_labels, yticklabels=iris_labels, xlabel="True Species", ylabel="Predicted Species")
```
